```python

from datetime import datetime

import pandas as pd
import numpy as np
```

```python
signUps = pd.read_csv('customer_signups.csv')

#Missing values
print(signUps.info())
```

```python
print(signUps.isnull().sum())
```

```python
#Cleaning empty values
signUps = signUps.dropna()
print(signUps.isnull().sum())
print(signUps.info())
```

```python
#Convert to date time on cleaned data set
#If the value is a string it will change that value to NaT (Not a Time)
signUps["signup_date"] = pd.to_datetime(signUps["signup_date"], errors='coerce')
print(signUps.signup_date)
```

```python
print(signUps.iloc[79])
```

```python
print(signUps.gender)
```

```python
signUps['plan_selected'] = signUps['plan_selected'].str.title()
print(signUps['plan_selected'].unique())

signUps['plan_selected'] = signUps['plan_selected'].replace('Unknownplan', np.nan)
signUps['plan_selected'] = signUps['plan_selected'].replace('Prem', 'Premium')
print(signUps['plan_selected'].unique())
```

```python
print(signUps['gender'].unique())
```

```python
signUps['gender'] = signUps['gender'].str.title()
print(signUps['gender'].unique())
signUps['gender'] = signUps['gender'].replace('123', np.nan)
print(signUps['gender'].unique())
print(signUps['gender'].isnull().sum())
```

```python
print(signUps['age'].unique())
signUps['age'] = signUps['age'].replace(['unknown', np.nan, '206'], np.nan)
print(signUps['age'].unique())
signUps['age'] = signUps['age'].replace('thirty', '30')
print(signUps['age'].unique())
```

```python
signUps['source'] = signUps['source'].replace('??', np.nan)
signUps['marketing_opt_in'] = signUps['marketing_opt_in'].replace(['Nil', 'None'], 'No')

print(signUps['source'].unique())
print(signUps['region'].unique())
print(signUps['plan_selected'].unique())
print(signUps['gender'].unique())
print(signUps['age'].unique())
print(signUps['marketing_opt_in'].unique())
```

```python
signUpsNoDupes = signUps.drop_duplicates(subset='customer_id')
print(signUps.duplicated(subset='customer_id'))
print(signUpsNoDupes.duplicated())
```

```python
print(signUpsNoDupes.isnull().sum())
```

```python
cleanSet = signUpsNoDupes.dropna()
print(cleanSet.isnull().sum())
```

Task 3: Summary Outputs

```python
cleanSet['signup_date'] = pd.to_datetime(cleanSet['signup_date'], errors='coerce')
signUpsPerWeek = cleanSet.groupby(pd.Grouper(key='signup_date', freq='W'))['customer_id'].count()
print(signUpsPerWeek.describe())
```

```python
source = cleanSet.groupby('source')['customer_id']
print(source.describe())
```

```python
region = cleanSet.groupby('region')['customer_id']
print(region.describe())
```

```python
plan = cleanSet.groupby('plan_selected')['customer_id']
print(plan.describe())
```

```python
opt_ins = cleanSet.groupby('gender')['marketing_opt_in'].value_counts()
print(opt_ins)
```

```python
ages = cleanSet.groupby('age')['age'].count()
print(ages.describe())
print("Null ages: ", cleanSet['age'].isnull().sum())
```

Task 5: Support Tickets

```python
supportTickets = pd.read_csv('support_tickets.csv')
mergedSet = pd.merge(cleanSet, supportTickets, on='customer_id')
mergedSet['ticket_date'] = pd.to_datetime(mergedSet['ticket_date'], errors='coerce')
filterByContact = mergedSet.where((mergedSet['ticket_date']-mergedSet['signup_date']).dt.days <= 14)
filterByContact.dropna()
print(filterByContact.nunique())
```

```python
activity = mergedSet.groupby(['plan_selected', 'region'])['issue_type'].value_counts()
print(activity)
```
